```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("ht_data.csv")
```

```python
df.head(5)
df.isnull()
df1=df.dropna(axis=1,how='any')
df1.describe()
```

```python
x=df1.iloc[:,1:12].values
```

```python
x[0:10]
```

```python
y=df1.target.values
```

```python
y[0:10]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn import svm
```

```python
sv=svm.LinearSVC()
sv.fit(xtrain,ytrain)
```

```python
ypred=sv.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```

Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(xtrain,ytrain)
```

```python
ypred1=rf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(ytest,ypred1)
```

```python
accuracy_score(ytest,ypred1)
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred2=lr.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(ytest,ypred2)
```

```python
accuracy_score(ytest,ypred2)
```

DT

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',max_depth=5)
```

```python
dt.fit(xtrain,ytrain)
```

```python
ypred3=dt.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(ytest,ypred3)
```

```python
accuracy_score(ytest,ypred3)
```
